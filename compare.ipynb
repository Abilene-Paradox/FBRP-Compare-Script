```python
import pandas as pd
import numpy as np
```

```python
# read the files
df1 = pd.read_excel("Rodney Key 1.xlsx")
df2 = pd.read_excel("Hillary Key 2.xlsx")
```

```python
# checl the shape
print("df1 shape", df1.shape)
print("df2 shape", df2.shape)
```

```python
df1["Number"] = df1.index
df2["Number"] = df2.index
```

```python
# Remove the index columns
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
# check the shape of df1
df1.shape
```

```python
# check the shape of df2
df2.shape
```

```python
# Highlighting the difference between two DataFrames
df_all = pd.concat([df1.set_index("Number"), df2.set_index("Number")], axis="columns", keys=["Rodney_Key1", "Hillary_Key2"])
df_final = df_all.swaplevel(axis="columns")[df1.columns.drop('Number')[:]]
df_final
```

```python
# pip install --upgrade pandas
```

```python
def highlight_diff(data, color='red'):
    attr = 'background-color: {}'.format(color)
    other = data.xs('Rodney_Key1', axis='columns', level=-1)
    return pd.DataFrame(np.where(data.ne(other, level=0), attr, ''),
                        index=data.index, columns=data.columns)

x = df_final.style.apply(highlight_diff, axis=None)
# print(type(x))
x.to_excel("Comparedrrr.xlsx")
```
